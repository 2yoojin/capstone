```python
import requests
import xml.etree.ElementTree as ET
import pandas as pd
from datetime import datetime, timedelta

observatory_codes = [1018640, 1018655, 1018658, 1018662, 1018669, 
                     1018670, 1018675, 1018680, 1018683, 1018692,
                     1018695, 1018697, 1018698, 1019630]
observatory_codes = [str(code) for code in observatory_codes]
```

```python
# 데이터를 저장할 빈 리스트 생성
all_data = []

# URL 기본 부분
base_url = 'https://api.hrfco.go.kr/52832662-D130-4239-9C5F-730AD3BE6BC6/waterlevel/list/1H/'

for code in observatory_codes:
    # 시작 날짜 설정
    start_date = datetime(2010, 1, 1, 0, 0)
    end_date = datetime(2010, 12, 31, 23, 0)    
    # 시작 날짜부터 끝 날짜까지 데이터 가져오기
    while start_date <= datetime(2023, 12, 31, 23, 0):
        # 날짜를 문자열로 변환하여 URL에 포함
        start_date_str = start_date.strftime('%Y%m%d%H')
        end_date_str = end_date.strftime('%Y%m%d%H')
        url = base_url + code + '/' + start_date_str + '/' + end_date_str + '.xml'
        print(code, start_date, end_date)
        # URL 불러오기
        response = requests.get(url)

        # XML 파싱
        tree = ET.ElementTree(ET.fromstring(response.content))

        # DataFrame으로 변환
        data = []
        for waterlevel in tree.iter('Waterlevel'):
            fw = waterlevel.findtext('fw').strip() if waterlevel.find('fw') is not None else None
            wl = waterlevel.findtext('wl').strip() if waterlevel.find('wl') is not None else None
            wlobscd = waterlevel.findtext('wlobscd').strip() if waterlevel.find('wlobscd') is not None else None
            ymdhm = waterlevel.findtext('ymdhm').strip() if waterlevel.find('ymdhm') is not None else None

            data.append({'fw': fw, 'wl': wl, 'wlobscd': wlobscd, 'ymdhm': ymdhm})

        # 전체 데이터 리스트에 추가
        all_data.extend(data)

        # 다음 연도로 이동
        start_date = datetime(start_date.year + 1, 1, 1, 0, 0)
        end_date = datetime(end_date.year + 1, 12, 31, 23, 0)

# DataFrame 생성
water_level_df = pd.DataFrame(all_data)
water_level_df
```

```python
import requests

def download_file(file_url, save_path):
    try:
        response = requests.get(file_url)
        
        if response.status_code == 200:
            with open(save_path, 'wb') as f:
                f.write(response.content)
            print(f"File downloaded successfully and saved to {save_path}")
            
            # 다운로드된 파일 내용 확인
            with open(save_path, 'rb') as f:
                file_content = f.read(100)  # 처음 100바이트만 읽기
            print("First 100 bytes of the file:")
            print(file_content)
            
        else:
            print(f"Failed to download file. Status code: {response.status_code}")
    except requests.exceptions.RequestException as e:
        print(f"An error occurred: {e}")

# URL과 저장 경로 변수를 지정합니다.
url = 'https://apihub.kma.go.kr/api/typ01/url/kma_sfctm2.php?tm=2010010100&stn=0&help=1&authKey=oChwlojFS7aocJaIxcu2ow'
save_file_path = '1111.zip'

# 파일 다운로드 함수를 호출합니다.
download_file(url, save_file_path)
```
